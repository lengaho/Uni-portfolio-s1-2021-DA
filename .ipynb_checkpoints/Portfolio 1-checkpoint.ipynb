# Analysis of Cycling Data

## Loading Data

The first dataset is an export of the ride data from [Strava](https://strava.com/), an online social 
network site for cycling and other sports.  This data is a log of every ride since the start of 2018
and contains summary data like the distance and average speed.  It was exported using 
the script `stravaget.py` which uses the stravalib module to read data. Some details of
the fields exported by that script can be seen in [the documentation for stravalib](https://pythonhosted.org/stravalib/api.html#stravalib.model.Activity). 

The exported data is a CSV file so that's easy to read, however the date information in the file is 
recorded in a different timezone (UTC) so we need to do a bit of conversion.  In reading the data I'm
setting the index of the data frame to be the datetime of the ride. 

If the below cell gives error in `tz_localize`, use `tz_convert` as Pandas is time zone aware. 

```python
strava = pd.read_csv('data/strava_export.csv', index_col='date', parse_dates=True)
strava.index = strava.index.tz_convert('UTC')
strava.head()
```

The second dataset comes from an application called [GoldenCheetah](https://www.goldencheetah.org/) which provides
some analytics services over ride data.  This has some of the same fields but adds a lot of analysis of the 
power, speed and heart rate data in each ride.  This data overlaps with the Strava data but doesn't include all 
of the same rides. 

Again we create an index using the datetime for each ride, this time combining two columns in the data (date and time) 
and localising to Sydney so that the times match those for the Strava data. 

```python
cheetah = pd.read_csv('data/cheetah.csv', skipinitialspace=True)
cheetah.index = pd.to_datetime(cheetah['date'] + ' ' + cheetah['time'])
cheetah.index = cheetah.index.tz_localize('Australia/Sydney')
cheetah.head()
```

The GoldenCheetah data contains many many variables (columns) and we won't go into all of them here. Some
that are of particular interest for the analysis below are:

Here are definitions of some of the more important fields in the data. Capitalised fields come from the GoldenCheetah data
while lowercase_fields come from Strava. There are many cases where fields are duplicated and in this case the values
should be the same, although there is room for variation as the algorithm used to calculate them could be different
in each case. 

  * Duration - overall duration of the ride, should be same as elapsed_time
  * Time Moving - time spent moving (not resting or waiting at lights), should be the same as moving_time
  * Elevation Gain - metres climbed over the ride
  * Average Speed - over the ride
  * Average Power - average power in watts as measured by a power meter, relates to how much effort is being put in to the ride, should be the same as  * average_watts' from Strava
  * Nonzero Average Power - same as Average Power but excludes times when power is zero from the average
  * Average Heart Rate - should be the same as average_heartrate
  * Average Cadence - cadence is the rotations per minute of the pedals
  * Average Temp - temperature in the environment as measured by the bike computer (should be same as average_temp)
  * VAM - average ascent speed - speed up hills
  * Calories (HR) - Calorie expendature as estimated from heart rate data
  * 1 sec Peak Power - this and other  'Peak Power' measures give the maximum power output in the ride over this time period.  Will be higher for shorter periods. High values in short periods would come from a very 'punchy' ride with sprints for example.
  * 1 min Peak Hr - a similar measure relating to Heart Rate
  * NP - Normalised Power, a smoothed average power measurement, generally higher than Average Power 
  * TSS - Training Stress Score, a measure of how hard a ride this was
  * device_watts - True if the power (watts) measures were from a power meter, False if they were estimated
  * distance - distance travelled in Km
  * kudos - likes from other Strava users (social network)
  * workout_type - one of  'Race',  'Workout' or  'Ride'
  
  
Some of the GoldenCheetah parameters are defined [in their documentation](https://github.com/GoldenCheetah/GoldenCheetah/wiki/UG_Glossary).  

## Combining Data Frames

The first task is to combine these two data frames using the `join` [method of Pandas](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html#joining-on-index).   The goal is to keep only those rows of data 
that appear in __both__ data frames so that we have complete data for every row.  

```python
# Join the two data frames
left = strava
right = cheetah
rides_inner = left.join(right, how="inner")
rides_inner.shape
```

```python
# Variables (columns) of interest - including duplicated fields
rides_raw = rides_inner[['Duration', 'elapsed_time', 'Time Moving', 'moving_time', 'Elevation Gain', 'elevation_gain', 'Average Speed', 
                       'Average Power', 'average_watts', 'Nonzero Average Power', 'Average Heart Rate', 'average_heartrate',
                       'Average Cadence', 'Average Temp', 'average_temp', 'VAM', 'Calories (HR)', '1 sec Peak Power',
                       '1 min Peak Hr', 'NP', 'TSS', 'device_watts', 'distance', 'kudos', 'workout_type']]
print(rides_raw.shape)
rides_raw.head()
```

**Note:** Values of the dataset will be picked from Cheetah, rather than Strava, for the following duplicated fields:
* Duration (sec) ~ elapsed_time (min)
* Time Moving (sec) ~ moving_time (min)
* Average Power (watts) ~ average_watts (watts)
* Elevation Gain (m) ~ elevation_gain (m)  (Elevation Gain > elevation_gain)
* Average Heart Rate ~ average_heartrate
* Average Temp ~ average_temp

```python
# Variables (columns) of interest - removing duplicated fields
rides_raw2 = rides_inner[['time', 'Duration', 'Time Moving', 'Elevation Gain', 'Average Speed', 'Average Power', 
                        'Nonzero Average Power', 'Average Heart Rate', 'Average Cadence', 'Average Temp', 
                        'VAM', 'Calories (HR)', '1 sec Peak Power', '1 min Peak Hr', 'NP', 'TSS', 'device_watts', 
                        'distance', 'kudos', 'workout_type'
                      ]]
print(rides_raw2.shape)
rides_raw2.head()
```

```python
# Remove rides with no measured power (where device_watts is False) - these are commutes or MTB rides
rides = rides_raw2[rides_raw.device_watts == True]

print(rides_raw2.shape)
print(rides.shape)
rides.head()
```

### Part 1.2. Relationships between variables:
Variables of interest:
* Distance
* Moving Time
* Average Speed
* Heart Rate
* Power (watts)
* Normalised power (NP)
* Training Stress Score
* Elevation Gain

```python
# Plot scatter plots for different variables of interest
interest = rides[['Time Moving', 'distance', 'TSS', 'Elevation Gain', 
                 'Average Speed', 'Average Power', 'NP', 'Average Heart Rate']]

sns.pairplot(interest)
plt.suptitle('Scatter plots of variables of interest', y=0, fontsize=15)
plt.show()
```

**Comment:**
The scatter plots show that:
* There is a positive linear relationship between **Time Moving, Distance, TSS & Elevation Gain**. 

* There is a positive linear relationship between **Average Speed, Average Power & Average Heart Rate**.
    * Note: The heart rate dataset has some obvious outliers where heart rate is 0. These outliers might stem from human's or device's failure to record the actual heart rate. There is no other cases for a healthy and alive person to have zero average heart rates during sports activities.

* There is a positive linear relationship between **NP & Average Power**.

There are further analysis with correlation & possible explanations below.

```python
# Check correlations between variables of interest
interest.corr(method ='pearson') 
```

```python
# Plot a heatmap for correlation
corr = interest.corr(method ='pearson') 
sns.heatmap(data=corr, xticklabels=True, yticklabels=True, annot=True, linewidths=.1, cmap='mako')

plt.title('Heatmap of Correlation', fontsize=13)
plt.show()
```

**Comment:**
The correlation table & heatmap shows that:
* **Time Moving, Distance, TSS & Elevation Gain** - vary together positively (as their corrs are very close to 1)
    * *Time Moving & Distance* (corr = 0.933596 ~ 1): 
        * The more time was spent moving, the greater the distance travelled.
        * Possible explanation: The travelling distance would naturally increase as more time was spent moving.
    * *Distance & Elevation Gain* (corr = 0.922565 ~ 1):
        * The greater the distance travelled was, the greater the metres climbed over the ride were.   
        * Possible explanation: The amount of climbing would increase as the cycling path got longer.
    * *Training Stress Score (TSS)*:
        * The greater the metres climbed over the ride were, the harder the ride became. (corr = 0.828928)
        * The difficulty of the ride is greatly and positively influenced by the amount of climb, moving duration and distance of the ride. (corr = 0.828928, 0.811409 & 0.805468, respectively)
        * Possible explanation: People would normally find it harder to perform rides with more climbing amount, longer durations or longer distances.
        
* **Average Speed, Average Power & Average Heart Rate** - vary together positively 
    * *Average Speed & Average Power* (corr = 0.814403):
        * The greater the effort was being put to the ride, the higher the speed was.
        * Possible explanation: Similar to running, a rider would cycle faster with greater effort.
    * *Average Heart Rate, Average Speed & Average Power*:
        * The average heart rate has a **moderate** positive relationship with the average power (corr = 0.448912) and speed (corr = 0.422165).
        * Similar to running, the heart rate increased as more power and speed was put to the ride.

* **NP & Average Power** - vary together positively (corr = 0.844487)
    * Nonzero Average Power (NP) is the same as Average Power but excludes times when power is zero from average. This explains the positive relationship between NP and Average Power.

**Comments:**

* This variability in size is likely due to the nature of ride being less intense and more of a regular routine than race or workout.

### Part 1.4. Relationship between `rides` and `weather`: 
* `weather` data in 2018-2019 is from [Bureau of Meterology](http://www.bom.gov.au/)
* Temperature variable used is `Maximum temperature (Degree C)` column which represents maximum temperature for the day. 
* Relationships are examined between:
    * distance travelled per ride and the temperature
    * average speed of ride and the temperature

```python
weather2018 = pd.read_csv('data/weather_data_2018.csv', skipinitialspace=True)
weather2019 = pd.read_csv('data/weather_data_2019.csv', skipinitialspace=True)
print('weather2018: ', weather2018.shape)
print('weather2019: ', weather2019.shape)

# Concat 2 weather datasets
weather_total = pd.concat([weather2018, weather2019], ignore_index = True)
print('weather_total: ', weather_total.shape)

# Set Date index
weather_total.index = pd.to_datetime(weather_total[['Year', 'Month', 'Day']])
```

```python
# Weather dataset with variables of interest
weather = weather_total['Maximum temperature (Degree C)']
weather.head()
```

```python
# Split DateTime index into Date & Time MultiIndex
rides.index = pd.MultiIndex.from_arrays([rides.index.date, rides.index.time], names=['Date','Time'])
# Convert Time index to a normal column
rides.reset_index(level='Time', inplace = True) 
rides.head()
```

```python
# Join rides & weather datasets
left = rides
right = weather
ridesNew = left.join(right, how="inner")
ridesNew.head()
```

```python
# Check correlations between Temperature vs Distance & Average speed
ridesNew[['Maximum temperature (Degree C)', 'distance', 'Average Speed']].corr(method ='pearson') 
```

**Comment:**
The correlation table indicates that:
* There is almost no relationship between temperature and the distance travelled per ride as their correlation of -0.018843 is very close to 0.
* There is a very weak negative or almost no relationship between temperature and the average speed of ride as their correlation equals to -0.119986.

```python
# Create a scatter plot of Distance vs Temperature
sns.scatterplot(data=ridesNew, x='Maximum temperature (Degree C)', y='distance', color='orange', s = 30)
plt.xlabel('Maximum temperature (Degree C)')
plt.ylabel('Distance (km)')
plt.title('Scatter plot: Distance vs. Temperature', fontsize=13)
plt.show()

# Identify Distance = dependent variable y, Temperature = independent variable x
```

**Comment:**
According to the scatter plot, there is hardly any relationship between temperature and the distance travelled per ride.

```python
# Create a scatter plot of Average Speed vs Temperature
sns.scatterplot(data=ridesNew, x='Maximum temperature (Degree C)', y='Average Speed', color='blue', s = 30)
plt.xlabel('Maximum temperature (Degree C)')
plt.ylabel('Average Speed')
plt.title('Scatter plot: Average Speed vs. Temperature', fontsize=13)
plt.show()

# Identify Average Speed = dependent variable y, Temperature = independent variable x
```

**Comment:**
On the basis of the scatter plot, there is hardly any relationship between temperature and the average speed of ride.

## Conclusion

#### Patterns among variables
There is a very strong positive relationship between the moving time, distance, training stress score (TSS) and elevation gain. 
It is quite obvious that the distance travelled increases as more time is spent moving and the amount of climb is more likely to rise as the cycling path becomes longer. Consequently, the difficulty of the ride will escalate in line with the increment in the climbing, moving duration and distance of the ride. 

The similar pattern is also found in the association between the average speed, the average power and the average heart rate. Simply speaking, with more effort being put into the ride, the rider will cycle faster with a rising heart rate.

#### Trends among different activities - ride, race & workout
There is an imbalance in the size of three categories. While the numbers of race and workout are approximately the same, the number of rides is more than five times greater than either that of race or workout. This huge difference in occurrences is likely caused by the nature of ride being less intense and more of a regular routine than race or workout which drain more energy, and thus, may not be prolonged.

##### Ride:
Ride has the highest records of moving time, distance and elevation. A large number of the rides were extended over longer durations and greater distances. The majority of ride activities included climbing while some did not. Ride also has the overall lowest figures in the average speed, power and heart rate. All of these findings can be explained with the nature of ride being the most leisure activity, requiring much less speed and power than workout and race. Additionally, ride often has larger spreads in many variables due to having a much bigger sample size.  
    
##### Race:
Race involves less time and distance and has the highest records in the average speed, power and heart rate. Every race activity included climbing regardless of the climbing amount performed. The essence of race requires greatest effort, power and speed within a specified period of time and distance.
    
##### Workout:
Workout activities also record less time and distance and greater speed, power and heart rate than rides. Second to race, workout has the second highest median of average speed, power & heart rate. However, none of the workout activities involved climbing.
    
#### Little or no relationship between rides and weather
The findings indicate that there is almost no relationship between temperature with the distance travelled or the average speed of ride.
